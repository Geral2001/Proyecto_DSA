```python
import pandas as pd
```

```python
ic_ips=pd.read_csv("C:/Users/DARLAN/Documents/GitHub/Proyecto_DSA/data/raw/Indicadores_salud/Clicsalud-_Indicadores_de_calidad_IPS_20251018.csv", encoding="utf-8")
```

```python
ic_ips.head()
```

encontramos que hay departamentos con cod departamento -1, hacemos la respectiva limpieza para evitar problemas

```python
ic_ips=ic_ips[ic_ips["coddepartamento"]!=-1]
ic_ips.head()
```

separamos el año y el mes de la medición

```python
ic_ips["periodo"]=ic_ips["periodo"].astype(str)
ic_ips["anio"]=ic_ips["periodo"].str[:4]
ic_ips["mes"]=ic_ips["periodo"].str[4:]
```

```python
print(ic_ips[["periodo", "anio", "mes"]].head())
```

```python
import pandas as pd


cols = ["numerador", "denominador", "resultado"]
for c in cols:
    ic_ips[c] = (
        ic_ips[c]
        .astype(str)          
        .str.replace(",", "", regex=False)  
        .astype(float)         
    )

# Estadísticas descriptivas básicas
estadisticas = ic_ips[cols].describe().T  
print(estadisticas)


promedios = ic_ips[cols].mean()
medianas = ic_ips[cols].median()
desviaciones = ic_ips[cols].std()

print("\nPromedios:\n", promedios)
print("\nMedianas:\n", medianas)
print("\nDesviaciones estándar:\n", desviaciones)
```

se encontraron elementos interesantes, donde el denominador tiene una alta variabilidad, en comparación el numerador, lo que desde la intuición nos podría indicar que hay un desempeño bueno de las IPS en la gestión de servicios de salud, lo cual nos puede resultar de útilidad a la hora de evidenciar cambios en las métricas de bienestar o conformidad frente a la labor que desempeñan en el sistema de salud

```python
ic_ips.to_csv("data/processed/ic_ips.csv", index=False)
```

```python
pila=pd.read_csv("C:/Users/Asus/Documents/GitHub/Proyecto_DSA/data/raw/PILA/Planilla_Integrada_de_Liquidación_de_Aportes_-_PILA_20251018.csv")
```

```python
pila.head()
```

```python
pila.info()
```

```python
col = ["NumeroCotizantes", "NumeroCotizaciones"]

for c in col:
    pila[c] = (pila[c]
                .astype(str)
                .str.replace('\u00a0', '', regex=False)     
                .str.replace(r'\s+', '', regex=True)        
                .str.replace(',', '', regex=False))         

pila[col] = pila[col].apply(pd.to_numeric, errors="coerce")
```

```python
pila.info()
```

```python
cols = ["NumeroCotizantes", "NumeroCotizaciones"]
for c in cols:
    pila[c] = (
        pila[c]
        .astype(str)          
        .str.replace(",", "", regex=False)  
        .astype(float)         
    )

# Estadísticas descriptivas básicas
estadisticas = pila[cols].describe().T  
print(estadisticas)


promedios = pila[cols].mean()
medianas = pila[cols].median()
desviaciones = pila[cols].std()

print("\nPromedios:\n", promedios)
print("\nMedianas:\n", medianas)
print("\nDesviaciones estándar:\n", desviaciones)
```

hay una baja diferenciación entre las dos variables númericas de cotizantes y número de cotizaciones, lo cual valdría la pena explorar en mayor medida, lo que nos puede generar problemas de multicolinealidad en caso de la implementación de modelos de regresión, por lo que sería necesario incluir modelos que incluyan penalización para minimizar el riesgo de generar correlaciones espurias. 

```python
pila.to_csv("data/processed/pila.csv")
```
