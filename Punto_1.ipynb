```python
import pandas as pd
```

```python
ic_ips=pd.read_csv("C:/Users/DARLAN/Documents/GitHub/Proyecto_DSA/data/raw/Indicadores_salud/Clicsalud-_Indicadores_de_calidad_IPS_20251018.csv", encoding="utf-8")
```

encontramos que hay departamentos con cod departamento -1, hacemos la respectiva limpieza para evitar problemas

```python
ic_ips=ic_ips[ic_ips["coddepartamento"]!=-1]
ic_ips.head()
```

separamos el año y el mes de la medición

```python
ic_ips["periodo"]=ic_ips["periodo"].astype(str)
ic_ips["anio"]=ic_ips["periodo"].str[:4]
ic_ips["mes"]=ic_ips["periodo"].str[4:]
```

```python
print(ic_ips[["periodo", "anio", "mes"]].head())
```

```python
import pandas as pd


cols = ["numerador", "denominador", "resultado"]
for c in cols:
    ic_ips[c] = (
        ic_ips[c]
        .astype(str)          
        .str.replace(",", "", regex=False)  
        .astype(float)         
    )

# Estadísticas descriptivas básicas
estadisticas = ic_ips[cols].describe().T  
print(estadisticas)


promedios = ic_ips[cols].mean()
medianas = ic_ips[cols].median()
desviaciones = ic_ips[cols].std()

print("\nPromedios:\n", promedios)
print("\nMedianas:\n", medianas)
print("\nDesviaciones estándar:\n", desviaciones)
```

se encontraron elementos interesantes, donde el denominador tiene una alta variabilidad, en comparación el numerador, lo que desde la intuición nos podría indicar que hay un desempeño bueno de las IPS en la gestión de servicios de salud, lo cual nos puede resultar de útilidad a la hora de evidenciar cambios en las métricas de bienestar o conformidad frente a la labor que desempeñan en el sistema de salud

```python
ic_ips.to_csv("data/processed/ic_ips.csv", index=False)
```

# subsidiado

```python
import pandas as pd
```

```python
subsidiado=pd.read_csv("data/raw/afiliados_subsidiado/Población_Base_de_Datos_Única_de_Afiliados_BDUA_del_régimen_subsidiado_20251018.csv")
```

```python
subsidiado.head()
```

```python
#limpiamos departamento y municipio para poder hacer el cruce
from unidecode import unidecode

subsidiado["Departamento"] = (
    subsidiado["Departamento"]
    .astype(str)                               # asegura que sean strings
    .str.strip()                               # quita espacios al inicio y final
    .apply(lambda x: unidecode(x))             # quita tildes y caracteres especiales
    .str.replace(r"\s+", "", regex=True)  # colapsa múltiples espacios en uno solo
)

```

```python
subsidiado["Municipio"] = (
    subsidiado["Municipio"]
    .astype(str)                               # asegura que sean strings
    .str.strip()                               # quita espacios al inicio y final
    .apply(lambda x: unidecode(x))             # quita tildes y caracteres especiales
    .str.replace(r"\s+", "", regex=True)  # colapsa múltiples espacios en uno solo
)
```

```python
subsidiado["key"]=subsidiado["Departamento"]+subsidiado["Municipio"]
```

```python
#pegamos divipola
divipola=pd.read_excel("data/interim/DIVIPOLA_Municipios.xlsx")
```

```python
divipola.head()
```

```python
divipola["departamento"] = (
    divipola["departamento"]
    .astype(str)                               # asegura que sean strings
    .str.strip()                               # quita espacios al inicio y final
    .apply(lambda x: unidecode(x))             # quita tildes y caracteres especiales
    .str.replace(r"\s+", "", regex=True)  # colapsa múltiples espacios en uno solo
)
```

```python
divipola["municipio"] = (
    divipola["municipio"]
    .astype(str)                               # asegura que sean strings
    .str.strip()                               # quita espacios al inicio y final
    .apply(lambda x: unidecode(x))             # quita tildes y caracteres especiales
    .str.replace(r"\s+", "", regex=True)  # colapsa múltiples espacios en uno solo
)
```

```python
divipola["key"]=divipola["departamento"]+divipola["municipio"]
```

```python
subsidiado_final=pd.merge(subsidiado, divipola, how="left", on="key")
```

```python
subsidiado_final.head()
```

```python
subsidiado_final.info()
```

```python
#verificamos NA
print(subsidiado_final["cod_muni"].isna().sum())
```

```python
subsidiado_final_na=subsidiado_final[subsidiado_final["cod_muni"].isna()]
```

```python
subsidiado_final_na.head()
```

La base de subsidiados tiene mal escritos algunos municipios del país, es necesario hacer una corrección detallada de los municipios que hay que ajustar para evitar inconvenientes en los cruces de información que se realicen

```python
subsidiado_final.describe()
```

la única información númerica que se puede hacer descriptivas es cantidad, que tiene un muy alta desviación estándar, lo cual sugiere que pueden haber dinámicas urbanas afectando la variable, donde como es de esperarse, municipios de mayor categoría pueden arrastrar la distribución de la variable. 

```python
subsidiado_final.to_csv("data/processed/subsidiado.csv")
```
